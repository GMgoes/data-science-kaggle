```python
from collections import Counter
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import pandas as pd
import numpy as np
import os
print('OK');
```

```python
# Getting the dataset
df = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv', sep = ','); # nrows = 284807 (We define how many tuples we want, default = All)
classes = sorted(Counter(df['Class']).items());
print('OK');
```

```python
#Parallel Cordinate visualization
fig = px.parallel_coordinates(df, color = 'Class', dimensions = df.columns.tolist(),
                             color_continuous_scale = px.colors.diverging.Tropic,
                             color_continuous_midpoint = 0.5);
fig.show();
```

```python
# Histogram plot visualization
df['Class'].plot.hist(); # here define the attributes that we want to visualize in our histogram
```

```python
# Boxplot visualization
sns.boxplot(data = df, x = df['V1']); #data = df, x = df['V1'], y = df['Class'] TODO: check why this code is crashing
```

```python
#Violin plot visualization
attributes = df.columns.tolist();
attributes.remove('Time');
attributes.remove('Amount');
attributes.remove('Class');
sns.violinplot(df[attributes]) # here we can define which attributes we want to visualize, we can pass all 28 attributes
```

```python
# Z-score plot visualization
sns.histplot(data = df, x = 'V1', binwidth = 3, kde = True); #bins = 10
```

```python
#Heatpmap plot visualization
data = df.corr();
sns.heatmap(data);
```

```python
#Scatter plot visualization
sns.scatterplot(data = df, x = 'V1', y = 'V2', hue = 'Class');
```
