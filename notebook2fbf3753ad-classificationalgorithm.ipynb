```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score
from sklearn import tree
import pandas as pd
import numpy as np
import random
import os
print('OK');
```

```python
df = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv', sep = ',');
y = df['Class'];
y = y.to_numpy();
X = df.drop(columns = 'Class');
X = X.to_numpy();
print('OK');
```

```python
sm = SMOTE(random_state = 42, sampling_strategy = 'minority');
# sm = SMOTE(random_state = 42, sampling_strategy = 'not minority');
# sm = SMOTE(random_state = 42, sampling_strategy = 'not majority');
# sm = SMOTE(random_state = 42, sampling_strategy = 'all');
X_smote, y_smote = sm.fit_resample(X, y);
print('OK');
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, test_size = 0.2, random_state = 42);
print('OK');
```

```python
clf = tree.DecisionTreeClassifier(criterion = "log_loss", max_depth = 3);
print('OK');
```

```python
clf = clf.fit(X_train, y_train);
print('OK');
```

```python
tree.plot_tree(clf);
print('OK');
```

```python
predict = clf.predict(X_test)
accuracy = accuracy_score(predict, y_test)
print(f'Accuracy: {accuracy}')
```
